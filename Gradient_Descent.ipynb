```python
import pandas as pd
import numpy as np
```

```python
class GDRegressor:

  # CONSTRUCTOR
  def __init__(self,m,b,eta,epochs):
    self.m=m
    self.b=b
    self.eta=eta
    self.epochs=epochs

  # TRAINING
  def fit(self,x,y):
    for i in range(self.epochs):
      p=-2*(np.sum(y-self.m*x-self.b)*x)
      self.m=self.m-(self.eta*p)
      q=-2*(np.sum(y-self.m*x-self.b))
      self.b=self.b-(self.eta*q)

  # PREDICT
  def predict(self,x,y):
    return self.m*x+self.b
```

```python
model=GDRegressor(m=-100,b=50,eta=0.01,epochs=1000)
```
